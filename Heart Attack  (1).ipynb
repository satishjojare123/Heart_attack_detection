```python
import neurokit2 as nk
import matplotlib.pyplot as plt
import seaborn as sns
'''
This function returns r peak index of the data given
the data will be in array form which is the signal for which R peak needs to be found
'''
def get_r_peaks(signal_input,visualize=False):
    ecg_signal = signal_input
    _, rpeaks = nk.ecg_peaks(ecg_signal, sampling_rate=SAMPLING_RATE)
    # Visualize R-peaks in ECG signal
    if(visualize):
        plt.rcParams['figure.figsize'] = [8, 5]  # Bigger images
        plot = nk.events_plot(rpeaks['ECG_R_Peaks'], ecg_signal)
        # Zooming into the first 5 R-peaks
        plot = nk.events_plot(rpeaks['ECG_R_Peaks'][0:5], ecg_signal[0:300])
    rpeaks = rpeaks['ECG_R_Peaks']
    #print(rpeaks)
    filtered_peaks = list(filter(lambda r: (r > 150 and r < 10000) , rpeaks))
    return filtered_peaks
```

```python
'''
Returns an dict of all features for a signal
'''
def get_qrst(signal_input,visualize=False):
    rpeaks = get_r_peaks(signal_input)
    _, waves_dwt = nk.ecg_delineate(signal_input, rpeaks, sampling_rate=SAMPLING_RATE, method="dwt", show=visualize, show_type='all')
    waves_dwt['ECG_R_Peaks'] = rpeaks
    #print(waves_dwt)
    return waves_dwt
```
